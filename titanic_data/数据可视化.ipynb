```python
sex = df.groupby('Sex')['Survived'].sum()
sex.plot.bar()
plt.title('survived_count')
plt.show()
```

```python
# bar plot

df.groupby(['Sex','Survived'])['Survived'].count().unstack().plot(kind='bar',stacked='True')
plt.title('survived_count')
plt.show()
```

```python
# line plot (Ordered & Unordered)

plt.title('survived_count')
fare_sur = df.groupby(['Fare'])['Survived'].value_counts().sort_values(ascending=False)
fare_sur.plot(grid=True)
plt.show()
```

```python
fare_sur = df.groupby(['Fare'])['Survived'].value_counts()
fare_sur.plot(grid=True)
plt.show()
```

```python
cls_sur = df.groupby(['Pclass','Survived'])['Survived'].count().unstack().plot(kind='bar',stacked='True')
plt.show()
```

```python
cls_sur = df.groupby(['Pclass','Survived'])['Survived'].count().unstack()
cls_sur.plot.bar()
```

```python
# People of Higher Fare & Higher Pclass are more likely to survive.
```

```python
fare_sur = df.groupby(['Age', 'Survived'])['Survived'].count().sort_index().unstack()
fare_sur.plot(grid = True)
plt.show()
```

```python
df.Age[df.Pclass == 1].plot(kind='kde')
df.Age[df.Pclass == 2].plot(kind='kde')
df.Age[df.Pclass == 3].plot(kind='kde')
plt.xlabel("age")
plt.legend((1,2,3),loc="best")
```
