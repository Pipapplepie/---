```python
'''
Compare .read_csv() and .read_table():
They both do the same exact thing, read in data from a text file.
The only difference is that read_csv defaults the delimiter to a comma,
while read_table uses tab '\t' as its default.
'''

df = pd.read_table('train.csv', delimiter = ',')
type(df)  # pandas.core.frame.DataFrame

df.head()
```

```python
# Get data as chunks
chunker = pd.read_csv("train.csv", chunksize=1000)
print(chunker)  # pandas.io.parsers.readers.TextFileReader object, NOT dataframes

for chunk in chunker:
    print(chunk)
```
