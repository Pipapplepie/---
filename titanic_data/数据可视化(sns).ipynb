```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('result.csv')
```

```python
sns.countplot(x="Pclass", hue="Survived", data=df)  # 好简单的方法
```

```python
facet = sns.FacetGrid(df, hue="Survived",aspect=3)
facet.map(sns.kdeplot,'Age',shade= True)  # Kernel Density Estimation
facet.set(xlim=(0, df['Age'].max()))
facet.add_legend()
```
