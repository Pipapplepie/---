```python
df_cleaned = df.Cabin.fillna('missing')
df_cleaned = df.Age.fillna(method="pad")
df_cleaned = df.Embarked.fillna(method="pad", limit=1)  # we can propagate non-NA values forward or backward
```

```python
# Duplicates

df_cleaned = df.drop_duplicates(subset = 'Ticket', keep = False)
```

```python
df_cleaned.to_csv('train_cleaned.csv')
```

```python
'''
Use cut when you need to segment and sort data values into bins.
This function is also useful for going from a continuous variable to a categorical variable.
For example, cut could convert ages to groups of age ranges. 
Supports binning into an equal number of bins, or a pre-specified array of bins.
'''
```

```python
df['Ageband'] = pd.cut(df['Age'], 5, ordered = True, labels = [1,2,3,4,5])
df.head()
df.to_csv('equal_age.csv')
```

```python
del df['Ageband']
df['Ageband'] = pd.cut(df.Age,[0,5,15,30,50,80], right = False, labels = [1,2,3,4,5])
df.to_csv('int_age.csv')
```

```python
# qcut: Discretize variable into equal-sized buckets based on rank or based on sample quantiles.
```

```python
del df['Ageband']
df['Ageband'] = pd.qcut(df.Age, [0, .1, .3, .5, .7, .9], labels = [1,2,3,4,5])
df.to_csv('quant.csv')
```

```python
# Replace string text: Sex, Cabin, Embarked

df['Sex'].unique()
```

```python
df['Sex_int'] = df['Sex'].replace(['male','female'],[1,2])
df.head()
```

```python
df['Embarked'].value_counts()
```

```python
df['Embarked_int'] = df['Embarked'].replace(['S','C','Q'],[1,2,3])
df.head()
```

```python
df['Cabin'].nunique()  # Too many
```

```python
label_dict = dict(zip(df['Cabin'].unique(), range(df['Cabin'].nunique())))
df['Cabin_int'] = df['Cabin'].map(label_dict)

df.head()
```

```python
# One-hot
# One-hot Encoder

for feat in ["Age", "Embarked"]:
    x = pd.get_dummies(df[feat], prefix=feat)
    df = pd.concat([df, x], axis=1)
    
df.head()       
```

```python
#  Extract message

df['Title'] = df.Name.str.extract('([A-Za-z]+)\.', expand=False)
df.head()
```

```python
df.to_csv('train1.csv')
```
